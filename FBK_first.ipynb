```python
import numpy as np
import scipy as sp
import pandas as pd
import random
import matplotlib.pyplot as plt
import sys
import os
import fit_library as ft
import open_data

%matplotlib nbagg
%load_ext autoreload
%autoreload 2
```

```python
def analyze(DATA,channel=1,Low_th=-100,High_th=10000):

    Q_fit = ft.gauss_fit()
    data_fit = DATA['Q'][((DATA['id']==channel) & (DATA['Q']>Low_th) & (DATA['Q']<High_th))]
    print data_fit.shape
    Q_fit(data_fit,'sqrt')
    fig = plt.figure()
    ax0 = fig.add_subplot(111)
    Q_fit.plot(axis = ax0, 
               title = "Channel 1 - QDC",
               xlabel = "Code (LSB)",ylabel = "Hits")
```
