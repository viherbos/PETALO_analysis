```python
import fit_library_n as ft_n
import fit_library as ft
import pandas as pd
import random
import matplotlib.pyplot as plt
import sys
import os
import open_data
import numpy as np

%matplotlib nbagg
%load_ext autoreload
%autoreload 2
```

```python
def analyze(DATA,channel=1,Low_th=-100,High_th=10000,bins='sqrt'):

    Q_fit = ft.gauss_fit()
    data_fit = DATA['Q'][((DATA['id']==channel) & (DATA['Q']>Low_th) & (DATA['Q']<High_th))]
    #print data_fit.shape
    Q_fit(data_fit,bins)
    fig = plt.figure()
    ax0 = fig.add_subplot(111)
    Q_fit.plot(axis = ax0, 
               title = "Channel 1 - QDC",
               xlabel = "Code (LSB)",ylabel = "Hits")
    return Q_fit.p0[1],Q_fit.p0[2]

def analyze2(DATA,channel=1,Low_th=-100,High_th=10000,bins='sqrt',mu1=168,mu2=168,sigma=2):

    Q_fit = ft.gauss_fit2()
    data_fit = DATA['Q'][((DATA['id']==channel) & (DATA['Q']>Low_th) & (DATA['Q']<High_th))]
    #print data_fit.shape
    Q_fit(data_fit,[mu1,mu2,sigma,sigma],bins)
    fig = plt.figure()
    ax0 = fig.add_subplot(111)
    Q_fit.plot(axis = ax0, 
               title = "Channel 1 - QDC",
               xlabel = "Code (LSB)",ylabel = "Hits")
    return Q_fit.coeff[1],Q_fit.coeff[2]
```
